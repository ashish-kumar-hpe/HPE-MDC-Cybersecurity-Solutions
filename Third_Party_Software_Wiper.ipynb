```python
SSHFlow_Query = "MATCH (v0:Devices)-[edge:Netflow]->(v1:Devices) WHERE edge.dst_port=22 CREATE (v0)-[e:SSHFlow {epoch_time : toInteger(edge.epoch_time), duration : toInteger(edge.duration), protocol : toInteger(edge.protocol), src_port : toInteger(edge.src_port), dst_port : toInteger(edge.dst_port)}]->(v1) RETURN count(*)"

with neo4j_driver.session() as neo4j_Session:
    result1 = neo4j_Session.run(SSHFlow_Query)

print(result1.data())
```

```python
Final_Query = "MATCH (n1:Devices)-[r1:SSHFlow]->(n2:Devices) MATCH (n1)-[r2:SSHFlow]->(n2) MATCH (n2)-[rb:HostEvents]->(n2) WHERE rb.event_id=1100 AND n1 <> n2 AND r1 <> r2 AND r1.epoch_time <= r2.epoch_time AND r2.epoch_time - r1.epoch_time < 60 AND r2.epoch_time <= rb.epoch_time AND rb.epoch_time - r2.epoch_time < 180 RETURN n1,n2,r1.epoch_time,r2.epoch_time,rb.epoch_time limit 1000"


with neo4j_driver.session() as neo4j_Session:
    result2 = neo4j_Session.run(Final_Query)

print(result2.data())
```
